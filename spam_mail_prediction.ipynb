```python
# importring the dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# loding the dataset

mail_data=pd.read_csv('/content/mail_data (1).csv')
mail_data.head()
```

```python
# getting shape of the data

mail_data.shape
```

```python
# checking is there any null values

mail_data.isnull().sum()
```

LABEL ENCODING

```python
# labeling spam mail as 0,and ham mail as 1

mail_data.loc[mail_data['Category']=='spam','Category']=0
mail_data.loc[mail_data['Category']=='ham','Category']=1
mail_data.head()
```

```python
# seperating the data as text and label

x=mail_data['Message']
y=mail_data['Category']
print(x)
print(y)
```

```python
# train test split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# printing the train and test data shape

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

FEATURE EXTRACTION

```python
# transforming test data into features vector that can be used as the input for logistic regression

feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)
```

```python
# converting y_train and y_test as integer

y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_train_features)
```

TRAINING MODEL


```python
# traing and testing in train data

model=LogisticRegression()
model.fit(x_train_features,y_train)
x_pred=model.predict(x_train_features)
acc_train=accuracy_score(y_train,x_pred)
print(acc_train)
```

```python
# traing and testing in test data

y_pred=model.predict(x_test_features)
acc_test=accuracy_score(y_test,y_pred)
print(acc_test)
```

BUILDING A PREDCTIVE SYSTEM

```python
input_mail=["As per your request 'Melle Melle (Oru Minnaminunginte Nurungu Vettam)' has been set as your callertune for all Callers. Press *9 to copy your friends Callertune"]
```

```python
# convert text to fearturs vector

input_data_features=feature_extraction.transform(input_mail)

# making prediction

prediction=model.predict(input_data_features)
print(prediction)


if prediction[0]==1:
  print('HAM MAIL')
else:
  print('SPAM MAIL')
```
